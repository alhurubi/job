```python
import pandas as pd
import numpy as np
from datetime import datetime
#read the file
df = pd.read_csv('data.csv')
#except some row who still alive
df = df.drop([37,40,41,42,43,44,45], axis=0) 
df
```

```python
PRESIDENT = df.loc[:,'PRESIDENT']
year_of_birth = df.loc[:,'BIRTH DATE']
year_of_death = df.loc[:,'DEATH DATE']
```

```python
def convertor(month):
    months = ['Jan','Feb','Mar','Apr','May','June','July','Aug','Sep','Oct','Nov','Dec']
    return months.index(month)+1
```

```python
data = []
years_of_brith = []
for i in range(36):
    #name of PRESIDENT
    name = PRESIDENT[i]
    #birth
    birth = year_of_birth[i]
    #split the string to do months
    b = year_of_birth[i].split()
    d = year_of_death[i].split()
    #split for the month only
    b_m = year_of_birth[i].split()[0]
    d_m = year_of_death[i].split()[0]
    b_m = convertor(b_m)
    d_m = convertor(d_m)
    #day
    b_d = year_of_birth[i].split()[1]
    b_d = b_d[:-1] #take off extra commma
    d_d = year_of_death[i].split()[1]
    d_d = d_d[:-1]
    #year
    b_y = year_of_birth[i].split()[2]
    d_y = year_of_death[i].split()[2]
    #calc years
    result_years = int(d_y)-int(b_y)
    result_weeks = result_years*52
    result_days = result_weeks*7
    #add to the list
    data.append([name,birth,result_years,result_weeks,result_days])
    years_of_brith.append(b_y)
```

```python
table = pd.DataFrame(data,columns=('President', 'Birth Date', 'Age in Years', 'Age in Weeks', 'Age in Days'))
table
```

```python
#plot the new data
people = table.loc[:,'President'] + " " + years_of_brith
Birth_year = years_of_brith
brith_years = table.loc[:,'Age in Years']
```

```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcdefaults()
fig, ax = plt.subplots(figsize=(7, 7))
ax.barh(people, brith_years, align='center', color= "gray",height = 0.3) #Make a horizontal bar plot
ax.set_yticks(people, labels=people,fontsize=8)
ax.invert_yaxis()  
ax.set_xlabel('Age',fontsize=8,color="gray")
ax.set_title('Life Span of United States Presidents',fontsize=12,color="gray")
plt.show()
```
